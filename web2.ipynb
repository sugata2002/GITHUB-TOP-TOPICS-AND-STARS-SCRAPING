```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
main_link = "https://github.com/topics/3d"
```

```python
response = requests.get(main_link)
```

```python
response.status_code
```

```python
txt = response.text
```

```python
len(txt)
```

```python
with open ('web2.html' , 'w' , encoding="utf-8")  as f :
    f.write(txt)
```

```python
soup = BeautifulSoup(txt, 'html.parser')
```

```python
soup
```

```python
selecter = "text-bold wb-break-word"
auth_c_link = soup.find_all('a' ,{"class" : selecter})
```

```python
len(auth_c_link)
```

```python
auth_c_link 
```

```python
#links
base2_url =  "https://github.com"
link_list = []

for tag in auth_c_link :
    link_list.append(base2_url +   tag ["href"])

print (link_list)
```
